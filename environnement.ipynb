### Facteurs influençant l'adhésion à une idéologie

Concernant l'influence environnementale, culturelle et sociale sur l'acceptation d'une idéologie et sa croissance; nous prenons comme exemple le nationalisme mis en avant par Marine Le Pen en France et nous étudierons les divers facteurs contribuant dans l'adhésion à cette idéologie.

Le graphique ci-dessous représente le pourcentage du nombre de personnes adhérant à cette idéologie selon leur profession/classe sociale.

```python
import matplotlib.pyplot as plt

X=["Ouvrier","Employé","PI","Retraité","Cadre"]
Y=[0.56, 0.46, 0.33, 0.26, 0.18]
plt.plot(X,Y)
plt.xlabel("Profession")
plt.ylabel("Nombre d'adhérents au nationalisme")
plt.show()
```

PI: Profession Intermédiaire

On remarque bien que la profession et la classe sociale a une influence considérable sur l'adhésion à une idéologie.Ici, une majorité de la "classe populaire" choisit cette idéologie.

```python
X=["inférieur au Bac","Bac","Bac+2","Bac+3 et plus"]
Y=[0.45, 0.36, 0.31, 0.19]
plt.plot(X,Y)
plt.xlabel("Niveau d'étude")
plt.ylabel("Nombre d'adhérents au nationalisme")
plt.show()
```

D'après les données, on remarque une décroissance plus le diplôme est élevé.

On a representé ici la France sous forme d'un rectangle, avec les zones du nord et nord-est beaucoup plus adhérées à l'idéologie ainsi que les régions du sud. On peut voir dans le graphe du bas que les régions les plus rurales, adhèrent plus aux idéologies du front national que les régions de plus de 100.000 habitants. 
Mais pour expliquer l'anomalie observée dans quelques grandes villes, nous pouvons suggérer le facteur de l'immigration dans ces villes, qui pourrait être la cause de l'adhérence.

```python
X=["Rural","moins de 20K hbts","20 à 100K hbts","+ de 100K hbts/Paris"]
Y=[0.43, 0.35, 0.38, 0.28]
plt.plot(X,Y)
plt.xlabel("Catégorie d'agglomération")
plt.ylabel("Nombre d'adhérents au nationalisme")
plt.show()
```
